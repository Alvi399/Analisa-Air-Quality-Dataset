## Data Wrangling

### Gathering Data

```python
data1_df = pd.read_csv('PRSA_Data_Changping_20130301-20170228.csv')
data2_df = pd.read_csv('PRSA_Data_Aotizhongxin_20130301-20170228.csv')
data3_df = pd.read_csv('PRSA_Data_Dingling_20130301-20170228.csv')
data4_df = pd.read_csv('PRSA_Data_Dongsi_20130301-20170228.csv')
data5_df = pd.read_csv('PRSA_Data_Guanyuan_20130301-20170228.csv')
data6_df = pd.read_csv('PRSA_Data_Gucheng_20130301-20170228.csv')
data7_df = pd.read_csv('PRSA_Data_Huairou_20130301-20170228.csv')
data8_df = pd.read_csv('PRSA_Data_Nongzhanguan_20130301-20170228.csv')
data9_df = pd.read_csv('PRSA_Data_Shunyi_20130301-20170228.csv')
data10_df = pd.read_csv('PRSA_Data_Tiantan_20130301-20170228.csv')
data11_df = pd.read_csv('PRSA_Data_Wanliu_20130301-20170228.csv')
data12_df = pd.read_csv('PRSA_Data_Wanshouxigong_20130301-20170228.csv')
```

```python
all_df = pd.concat(
    [data1_df,
    data2_df,
    data3_df,
    data4_df,
    data5_df,
    data6_df,
    data7_df,
    data8_df,
    data9_df,
    data10_df,
    data11_df,
    data12_df],
    ignore_index= False
).reset_index()
all_df.head()    
```

### Cleaning Data

```python
# Misssing value
columns =["PM2.5","PM10","SO2","NO2","CO","O3","TEMP","PRES","DEWP","RAIN","WSPM"]
for c in columns:
    all_df[c].fillna(value=all_df[c].mean(),inplace= True)
all_df["wd"].fillna(value="unknown",inplace=True)
all_df.isna().sum()
```

```python
# all_df.drop(columns="index",axis=1,inplace=True)
all_df.describe(include="all")
```

## Exploratory Data Analysis (EDA)

### Explore All_df

```python
#Suhu rata - rata dari stiap wilayah
temp_mean_df = all_df.groupby(by="station").agg({
    "TEMP": "mean"
})
temp_mean_df.head()
```

```python
year_max_temp_df = all_df.groupby(by= "year").TEMP.max().sort_values(ascending=False)
year_max_temp_df.head()
```

```python
year_min_temp_df = all_df.groupby(by= "year").TEMP.min().sort_values(ascending=True)
year_min_temp_df.head()
```

```python
pm25_df = all_df.groupby(by="station").agg({
    "PM2.5": "max"
})
pm25_df.head()
```
